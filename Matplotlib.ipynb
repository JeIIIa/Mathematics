# Библиотека Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Визуализация данных при помощи matplotlib

**1. Постройте график функции y = x^2 для x в диапазоне от -10 до 10.**

```python
plt.figure(figsize=(6, 4)) # Создаем "холст" для графиков
x = np.arange(-10, 10.1, 0.1)  
y = x ** 2 # Каждому значению x соответствует свое значение y, в переменной y хранится массив
plt.plot(x, y)
plt.title('График функции y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

print(y)
```

**2. Постройте на одном рисунке графики функций y = x, y = x^2, y = x^3 для x в диапазоне [0, 5].**

```python
plt.figure(figsize=(6, 4))
x = np.arange(0, 5.1, 0.1)
plt.plot(x, x, label='y = x')
plt.plot(x, x**2, label='y = x^2')
plt.plot(x, x**3, label='y = x^3')
plt.title('Графики функций')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() # Подсказка: где какой график изображен
plt.grid(True) # Включить/выключить сетку
plt.show() # Отобразить график ч
```

**3. Сгенерируйте 1000 случайных чисел из нормального распределения и постройте их гистограмму.**

```python
plt.figure(figsize=(6, 4))
data = np.random.randn(1000000)
plt.hist(data, bins=30, edgecolor='black') # На вход поступает массив чисел, количество "столбцов", цвет границы столбцов
plt.title('Гистограмма нормального распределения')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()
```

**4. Постройте круговую диаграмму для данных: [25, 35, 20, 20] с метками ['A', 'B', 'C', 'D'].**

```python
plt.figure(figsize=(6, 4))
values = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']
plt.pie(values, labels=labels) # Cоздание кругового графика, на вход поступает числовые значения для каждого сектора, название секторов
plt.title('Круговая диаграмма')
plt.show()
```

**5. Сгенерируйте два массива из 100 случайных чисел (x и y) и постройте диаграмму рассеяния.**

```python
plt.figure(figsize=(6, 4))
x = np.random.randn(100) 
y = np.random.randn(100)
plt.scatter(x, y, alpha=0.5)
plt.title('Диаграмма рассеяния')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True) # Включить/выключить сетку
plt.show()
```

**6. Постройте график функции y = sin(x) для x в диапазоне [0, 4π]. Добавьте подписи осей, заголовок и легенду. Сделайте линию графика пунктирной и красного цвета.**

```python
plt.figure(figsize=(6, 4))
x = np.arange(0, 4*np.pi, 0.1)  # от 0 до 4π
y = np.sin(x)
plt.plot(x, y, 'r--', label='y = sin(x)')  # 'r--' = красный пунктир
plt.title('График функции y = sin(x)')
plt.xlabel('x (радианы)')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

**7. Создайте рисунок с 4 subplot'ами (2x2), на каждом из которых постройте разные графики (линейный, гистограмму, scatter, круговую диаграмму).**

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8)) # fig - весь график, axes - массив 2x2

# Создание линейного графика
x = np.arange(0, 10, 0.1)
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Линейный график')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')
axes[0, 0].grid(True)

# Создание гистограммы
data = np.random.randn(1000)
axes[0, 1].hist(data, bins=30, edgecolor='black')
axes[0, 1].set_title('Гистограмма')
axes[0, 1].set_xlabel('Значение')
axes[0, 1].set_ylabel('Частота')

# Создание диаграммы рассеяния
x = np.random.randn(100)
y = np.random.randn(100)
axes[1, 0].scatter(x, y, alpha=0.5)
axes[1, 0].set_title('Scatter plot')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
axes[1, 0].grid(True)

# Создание круговой диаграммы
values = [30, 25, 20, 25]
labels = ['A', 'B', 'C', 'D']
axes[1, 1].pie(values, labels=labels, autopct='%1.1f%%')
axes[1, 1].set_title('Круговая диаграмма')

plt.tight_layout()
plt.show()
```
